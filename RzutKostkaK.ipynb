```python
import random
```

```python
rzuty = [random.randint(1,6) for i in range(100_000)]
```

```python
print(rzuty[0:30])
```

```python
licznik = {}
for oczka in range(1,7):
  licznik[oczka] = rzuty.count(oczka)
```

```python
print(licznik)
```

```python
procenty = {}
for i in range(1,7):
  procent = (licznik[i] / len(rzuty) * 100)
  procenty[i] = round(procent, 3)
```

```python
print(procenty)
```
