```python
produkty_AGD = [ ("Laptop", 6000), ("Myszka", 200), ("Klawiatura", 450), ("Monitor", 1200), ("Pendrive", 50), ("Słuchawki", 750), ("Podkładka", 40) ]
```

```python
#Usuń wszystkie produkty tańsze niż 100 zł.
#Dla pozostałych produktów zwiększ cenę o 23% VAT.
#Posortuj produkty od najdroższego do najtańszego.
```

```python
temperatury = [23.5, 40.2, 15.8, 50.1, 37.7, 12.3, 45.9]
```

```python
#Wybierz tylko temperatury większe niż 30°C.
#Przelicz temperatury z Celsjusza na Fahrenheit:
#Posortuj temperatury rosnąco.
```

```python
# ZAD 1------------------------------------
```

```python
one = list(filter(lambda x: x[1] > 100, produkty_AGD))
```

```python
print(one)
```

```python
two = list(map(lambda x:( x[0], x[1] * 1.23), one))
print(two)
```

```python
three = sorted(two, key=lambda x: x[1], reverse=1)
```

```python
print(three)
```

```python
temperatury = [23.5, 40.2, 15.8, 50.1, 37.7, 12.3, 45.9]
```

```python
temp_1 = list(filter(lambda x: x > 30, temperatury))
print(temp_1)
```

```python
temp_2 = list(map(lambda x: x * 1.8 + 32, temp_1))
temp_2 = list(map(lambda x: round(x,2), temp_2))
print(temp_2)
```

```python
temp_3 = sorted(temp_2)
print(temp_3)
```
