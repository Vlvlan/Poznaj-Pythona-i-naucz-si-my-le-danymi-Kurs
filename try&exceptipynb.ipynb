```python
lista = [0,1,2]

try:
  2 / 0
  print(lista[1])

except IndexError:
  print("Podano zły index listy")

except ZeroDivisionError:
  print("Dzielenie przez zero")

else:
  print("Operacja nie zwórciła błędu")

finally:
  print("Finally")
```

```python
def logowanie(user_id):
  if user_id <= 0:

    raise ValueError("ID jest niepoprawne")
  return f"Udało się zalogowac {user_id}"

try:
  print(logowanie(1))
except ValueError:
  print("Nie udało się zalogować")
```
