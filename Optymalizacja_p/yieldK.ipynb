```python
def kwadrat_g(n):
  for i in range(n):
    yield i ** 2

print(list(kwadrat_g(10)))
```

```python
def fib(x):
  a, b = 0,1
  while a < x:
    yield a
    a,b = b, a + b

for liczba in fib(10):
  print(liczba, end=" ")

```

```python
def liczby():
  n = 1
  while True:
    yield n
    n += 1
```

```python
l = liczby()
print(next(l))
```

```python
print(next(l))
print(next(l))
print(next(l))
```

```python
def test():
  yield "1"
  yield "2"
  yield "3"

t = test()

print(next(t))
print(next(t))
print(next(t))
print(next(t, "Koniec"))
```
