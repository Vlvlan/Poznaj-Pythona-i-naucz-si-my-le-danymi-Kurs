```python
import time
```

```python
def dekorator(funkcja):
  def opakowanie():

    start = time.perf_counter()
    funkcja()
    koniec = time.perf_counter()
    return f"Czas wykonania: {koniec - start:.3f}"

  return opakowanie

```

```python
@dekorator
def hello():
  print("Hello world")
```

```python
hello()
```
