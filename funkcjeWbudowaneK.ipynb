```python
liczby = [5,12,6,8,10,100,28,67,1,33]
piwko_nie_mozna = ["żubr", "harnas", "beczkowe", "kasztelan"]

oceny = {
    "Paweł": 5,
    "Monika": 4,
    "Adam": 3
}
```

```python
sortowanie_liczb = sorted(liczby, reverse=True)
print(sortowanie_liczb)
```

```python
sortowanie_string = sorted(piwko_nie_mozna, key= lambda s: len(s))
print(sortowanie_string)
```

```python
sortowanie_string_2= sorted(piwko_nie_mozna, key= lambda s: s[-1])
print(sortowanie_string_2)
```

```python
sortowanie_oceny = sorted(oceny)
print(sortowanie_oceny)
```

```python
sortowanie_oceny = dict(sorted(oceny.items(), key=lambda element: element[1]))
print(sortowanie_oceny)
```

FILTER

```python
numbers = list(range(1,31))
poczta = ["nowak123@onet.eu", "pawełek67@wp.pl", "Marian123456", "@wp.pl"]
```

```python
parzyste = list(filter(lambda x: x % 2 == 0, numbers))
```

```python
print(parzyste)
```

```python
poczta_sort = list(filter(lambda x: len(x) > 7 and "@" in x, poczta))
print(poczta_sort)
```

MAP

```python
lista = [1.123, 4.216, 3.14159, 8.8124, 1002.1214]
```

```python
lista_round  = list(map(lambda x: round(x,2), lista))
print(lista_round)
```
