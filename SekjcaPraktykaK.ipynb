**LISTA ZAKUPÓW**

```python
# Napisz program w pythonie który będzie symulował liste zakupów.
# Użytkownik może dodawać,usuwać oraz wyświetlać elementy.
# Dodaj opcje wyjścia z programu.
```

```python
main_list = []

while True:

  print("""Jaką chcesz przeprowadzic operacje?
  Wpisz 1 żeby wyświetlic elementy listy
  Wpisz 2 żeby dodać element do listy
  Wpisz 3 żeby usunąć element z listy
  Wpisz 4 żeby opuscic program
  """)

  odp = input("Twoja odpowiedz: ")

  if not odp in ["1","2","3","4"]:
  #jeśli odp użytkownika nie będzie zgodna z podaną listą pętla się zresetuje

    print("\nPodano błędny znak\n")
    continue

  if odp == "1": # wyświetlanie wszystkich elementów listy

    for index, wartosc in enumerate(main_list):
      # funkcja enumerate pozwala nam wyświetlić index i wartość listy

      print(index+1, wartosc)

  if odp =="2": # Dodawanie elementów do listy

    new_element = ""

    while new_element != "1":
      # tak długo jak użytkownik nie wpisze 1 pętla będzie się wykonywać

      new_element = input("Podaj produkt do dodania, jeśli chcesz wrócic do menu wpisz 1: ")

      if new_element == "1":
        continue

      else:
        main_list.append(new_element)

    for index, wartosc in enumerate(main_list): # wyświetlenie listy
      print(index+1, wartosc)

  if odp == "3": # usuwanie elementów z listy

    usun = input("Podaj element do usunięcia: ")

    main_list.remove(usun)
    # niestety jeśli użytkownik poda element którego nie ma program przestanie działać, jesteś w wstanie to obsłużyć?


  if odp == "4":
    break


```

**ZGADNI LICZBE**

```python
# Napisz program "zgadnij liczbe"
# komputer losuje liczbe z przedzału od 0 do 100 (za pomocą biblioteki random)
# Użytkownik próbuje zgadnąć liczbe wpisująć swoją odpowiedz
# jeśli podana odp będzie większa niż wylosowana liczba program powinien poinformować o tym użytkownika, i w drugą strone jeśli odp będzie za niska
# Po zgadnięciu liczby program powinien się zakonczyc i wyświetlic liczbe prób
```

```python
import random
```

```python
print(random.randint(0,100))
# metoda randit losuje liczbe (int) z przedziału który podamy (start, stop)
```

```python
liczba_komputera = random.randint(0,100)

print("Witaj w grze zgadnij liczbe, twoim celem jest zgadnięcie wylosowanej liczby w przedziale od 0 do 100")

i = 0 # zmienna która zlicza próby użytkownika...mogła mieć lepszą nazwe

while True: # pętla wykonuje się dopóki użytkonik nie zgadnie liczby

  odp = int(input("\nPodaj swoją liczbe:   "))

  if odp == liczba_komputera:
    print(f"Brawo wygrałes, potrzebowałes tylko {i} prób")
    break

  elif odp > liczba_komputera:
    print("Twoja odpowiedz jest za duza")

  elif odp < liczba_komputera:
    print("Twoja odpowiedz jest za mała")

  i+=1
```

**GRA W OCZKO**

```python
#Celem gry jest uzyskanie sumy kart jak najbliższej 21, ale bez jej przekroczenia.
# gracz i krupier dostają po dwie losowe karty (1 karta krupiera jest odkryta)
# karty 2–10 mają swoją wartość liczbową,
# walet, dama i król są warte po 10 punktów,
# AS może być wart 1 lub 11 — wybiera się korzystniejszą wartość.
#-----------------------------------------------------------------
# Po otrzymaniu kart gracz może:
# dobrać kartę („hit”),
# zatrzymać się („stand”) - wtedy sprawdza się karty z krupierem.
# Jeśli suma kart przekroczy 21 - gracz przegrywa automatycznie („bust”).
#------------------------------------------------------------------------
# Krupier też może dobierać karty
# Jeśli krupier przekroczy 21, gracz wygrywa
# Jeśli ani gracz ani krupier nie przekroczyli 21 wygrywa ten z większą sumą punktów, w przypadku tej samej sumy jest remis
```

```python
import random

def sumowanie(lista):
  """przyjmuje jeden parametr który jest listą kart gracza lub krupiera i zamienia karte AS jeśli taka jest na wartości 1 lub 11"""

  p = 0 # pomocnica zmienna która będzie się zwiększać jeśli w liscie pojawi się str AS

  for i in lista:

    if i == "AS":
      lista.remove(i)
      p +=1

    while p > 0: # pętla wykona się tyle razy ile w liscie pojawił się AS

      if sum(lista) <= 10:

        lista.append(11)

      else:
        lista.append(1)
      p -=1

  return lista




deck = ["AS", 10,10,10] # główna lista która będzie przechowywać  52 karty
# na start dodajemy str AS i 3 wartości po 10 które zastępują dame, jopka i króla

for i in range(2,11): # wypełniamy liste wartościami od 2 do 10
  deck.append(i)

deck *= 4 # mnożymy wszystkie elementy przez 4 żeby mieć dokładnie 52 karty

# Tworzymy dwie listy które będą przechowywać karty gracza i krupiera
player = []
croupier = []

random.shuffle(deck) # mieszamy liste z kartami co ma symulować tasowanie kart

# gracz i krupier dostają po dwie karty
# dzięki metodzie pop odrazu dodawaną karte z listy deck
player.append(deck.pop())
player.append(deck.pop())

croupier.append(deck.pop())
croupier.append(deck.pop())

print("Karta krupiera:",croupier[0]) # wyświetlamy jedną karte krupiera

while True:

  print("Twoje karty to:", player)

  odp = input("Wpisz 1 żeby dobrać karte\nWpisz 2 żeby sprawdzić")

  if odp == "1":

    # dodajemy karte dla gracza i sprawdzamy czy ma asa
    player.append(deck.pop())
    player = sumowanie(player)

    if sum(player) > 21: # jeśli suma wartości z listy gracza przekroczyła 21 program się konczy
      print("Przekroczyłes 21 punktów, przegrałes")
      break

  elif odp == "2":

    croupier = sumowanie(croupier)

    while sum(croupier) < 15: # jeśli krupier ma poniżej 15 punktów dobiera karte/karty
      croupier.append(deck.pop())
      croupier = sumowanie(croupier)

    if sum(croupier) > 21: # jeśli krupier przekroczył 21 punktów gracz wygrywa i program się konczy
      print("Krupier przepalił, gratuluje wygranej!")
      break

    player = sumowanie(player)

    # teraz kiedy wiemy że gracz i krupier mają poniżej 21 punktów sprawdzamy który wygrywa

    if sum(player) > sum(croupier):
      print("Brawo wygrałes")
      print(f"Twoje karty {player}, twoje punkty: {sum(player)}, karty krupiera {croupier}, jego suma punktów: {sum(croupier)}")
      break

    elif sum(player) < sum(croupier):
      print("Masz mniej punktów, przegrałes")
      print(f"Twoje karty {player}, twoje punkty: {sum(player)}, karty krupiera {croupier}, jego suma punktów: {sum(croupier)}")
      break

    elif sum(player) == sum(croupier):
      print("Remis, mieliscie tyle samo punktów")
      print(f"Twoje karty {player}, twoje punkty: {sum(player)}, karty krupiera {croupier}, jego suma punktów: {sum(croupier)}")
      break


  else:
    print("Podano bledny znak, spróbuj ponownie")



```

**KALKULATOR BMI**

```python
waga = float(input("Podaj swoją wage "))
wzrost = float(input("Podaj swój wzrost w metrach "))

wzrost = wzrost **2
BMI = waga / wzrost

print("Twoje BMI wynosi ",round(BMI,2))

if BMI < 18.5:
  print("niedowaga")

elif BMI > 18.5 and BMI < 24.99:
  print("waga prawidłowa")

elif BMI > 24.99 and BMI < 30:
  print("nadwaga")

else:
  print("Otyłość")
```

**Ciąg Fibonnaciego**


```python
ciag_f = [0,1]

odp = int(input("Ile razy ciąg fibonnaciego ma się wykonać  "))
x = 0

while x <= odp:
  #ciag_f.append(ciag_f[-1] + ciag_f[-2])
  ciag_f.append(ciag_f[len(ciag_f) - 2] + ciag_f[len(ciag_f) - 1])
  x+=1

print(ciag_f[:odp])
```
