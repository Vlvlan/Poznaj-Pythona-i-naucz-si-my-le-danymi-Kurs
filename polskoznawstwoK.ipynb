```python
import random
import time
# program polskoznawstwo to symulator gry "milionerzy" w którym użytkownik dostaje pytanie, i następnie 4 odpowiedzi do wyboru
# jeśli udzielona odp będzie zgodna z tą podaną to gracz dostaje punkt.
# Na koncu program robi podsumowanie zdobytych punktów i wylicza przedział procentowy jak dobrze poszło
# Poniewaz jest to przedział nawet zdobycie wszyskich punktów nie gwarantuje 100%, tylko przedział jest losowany od 80 do 100
```

```python
pytania_lista = [
  # każdy pojedynczy index listy to cały obiekt z trescią pytania, odpowiedziami i wskazaniem na prawidłową odpowiedz
    {
      "tresc" : "Co to jest 1 maja?", # key - tresc przechowuje pytanie
      "odpowiedzi" : {  # zwykły string key - odpowiedzi przechowuje słownik
            # w słowniku są 4 elementy które przechowują odpowiedzi do pytania
          "A" : "Święto pracy",
          "B" : "Święto grilla",
          "C" : "Zwykły dzień",
          "D" : "Wtorek",
      },
      "poprawna" : "B", # na koncu mamy zwykły element który przechowuje prawidłową odpowiedz z powyżej podanych
      },

    {
      "tresc" : "Gdzie w kuchni trzyma się kosz na śmeci?",
      "odpowiedzi" : {
          "A" : "Pod stołem",
          "B" : "Obok zlewu",
          "C" : "Pod zlewem",
          "D" : "Na parapecie",
      },
      "poprawna" : "C",
      },

    {
      "tresc" : "Jaki Polski obiad jest najlepszy?",
      "odpowiedzi" : {
          "A" : "Pomidorowa",
          "B" : "Kanapki",
          "C" : "Psia karma",
          "D" : "Schabowy z ziemniaczkami i mizerią",
      },
      "poprawna" : "D",
      },

    {
      "tresc" : "Czy miasto Łódź zostało zniszczone przy użyciu broni nukrearnej?",
      "odpowiedzi" : {
          "A" : "NIE",
          "B" : "NIE, chociaż tak wygląda",
          "C" : "TAK",
          "D" : "Łódź to piękne miasto z bogatą historią",
      },
      "poprawna" : "B",
      },
]
```

```python
losowanie = [0,1,2,3] # ta lista pomoże nam wybierać pojedynczo elementy z listy pytan
random.shuffle(losowanie) # metoda z bib.random która miesza losowo elementy podanego obiektu
SCORE = 1 # zmienna int która będzie zliczać punkty gracza

print("="*20+"polskoznawstwo".upper()+"="*20+"\n\n\n") # super ładny pasek

for i in range(len(pytania_lista)): # w pętli for działa cały program
# pętla wykona się tyle razy ile jest elementów w "pytania_lista"

  pytanie = pytania_lista[losowanie.pop()] # losujemy pojedynczy element

  print(f"\nPytanie: {pytanie["tresc"]}\n")
  #wyświetlamy pytanie za pomocą key - tresc

  for litera, tekst in pytanie["odpowiedzi"].items(): # wyświetlamy odpowiedzi
    print(f"   {litera}: {tekst}")

  while True: # w pętli while czekamy na odpowiedz użytkonika

    odpowiedz = input("Twoja odpowiedz (A/B/C/D): ").upper()

    if odpowiedz in "ABCD" and len(odpowiedz) < 2: # tylko jeśli użytkonik poda odpowiedz jako A,B,C lub D może wyjsc z pętli
      break
    else:
      print("Nieprawdiłowy wybór. Wpisz A, B, C lub D")

  if odpowiedz == pytanie["poprawna"]: # jeśli udzielona odpowiedz będzie zgodna z podaną w key poprawna gracz zyskuje punkt
    SCORE += 1


print("\n  ŁADOWANIE") # SUPER PIĘKNY PASEK ŁADOWANIA
for i in range(11):
  print("|"+"*" * i + " " * (10 - i) + "|")
  time.sleep(0.5) # metoda sleep z bib.time pozwala opóżnić wykonanie kodu na pół sekundy

p = random.randint((SCORE - 1) * 20 ,SCORE * 20) # wyliczamy przedział procentowy w zaleznosci od zdobytyk punktów (najgorszy przedział to od 0 do 20%)
print(f"\nTwój poziom polskośći to:{p}%") # wyświetlamy wynik koncowy

```
