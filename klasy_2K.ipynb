```python
class Temperatura:

  def __init__(self, stopnie):

    self.stopnie = stopnie


  def __str__(self):
    return f"{self.stopnie}°C"

  def __repr__(self):
    return f"Temperatura(stopnie={self.stopnie})"


```

```python
day_1 = Temperatura(30)
```

```python
print(day_1)
```

```python
print(repr(day_1))
```
