```python
import tkinter as tk

okno = tk.Tk()
okno.title("Kalkulator")
okno.geometry("400x500")
okno.resizable(False, False)

wyswietlacz = tk.Entry(okno, font=("Arial", 20), borderwidth=5, justify="right")
wyswietlacz.grid(row=0, column=0, columnspan=4)

def dodaj_liczbe(liczba):
    wyswietlacz.insert(tk.END, liczba)

def clean():
    wyswietlacz.delete(0, tk.END)

def oblicz():
    try:
        wynik = eval(wyswietlacz.get())
        wyswietlacz.delete(0, tk.END)
        wyswietlacz.insert(tk.END, str(wynik))
    except (SyntaxError, ZeroDivisionError, NameError):
        wyswietlacz.delete(0, tk.END)
        wyswietlacz.insert(tk.END, "Błąd")


buttons = [("7",1,0), ("8",1,1), ("9",1,2), ("/",1,3),
           ("4",2,0),("5",2,1),("6",2,2),("*",2,3),
           ("1",3,0),("2",3,1),("3",3,2),("-",3,3),
           ("0",4,0),("+",4,1), ("C",4,2), ("=",4,3),]

for (tekst, wiersz, kolumna) in buttons:

    if tekst == "C":
        run = clean
    elif tekst == "=":
        run = oblicz
    else:
        run = lambda t=tekst:dodaj_liczbe(t)


    tk.Button(okno, text=tekst, font=("Arial", 18), width=4, height=2, command=run).grid(row=wiersz,column=kolumna, padx=1,pady=1)




okno.mainloop()
```
