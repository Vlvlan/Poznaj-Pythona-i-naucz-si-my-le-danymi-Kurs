```python
class Bank:

  def __init__(self, user, __stan_konta):
    self.user =user
    self._saldo = __stan_konta

  @property
  def saldo(self):
    return self._saldo

  @saldo.setter
  def saldo(self, kwota):
    if kwota > 0:
      self._saldo = kwota
    else:
      raise ValueError
```

```python
class Zwierze:
  def __init__(self, gatunek):
    self.gatunek = gatunek

  def show(self):
    print(f"Jestem z gatunku {self.gatunek}")

```

```python
class Pies(Zwierze):
  def __init__(self, gatunek, imie):
    super().__init__(gatunek)
    self.imie = imie

  def przedstaw(self):
    print(f"Zwierze z gatunku {self.gatunek}, jego imie to {self.imie}")
```

```python
dog = Pies("wilczur", "Kostek")
```

```python
dog.przedstaw()
```

```python
class A:
  def witaj(self):
    print("Witaj A")

class B(A):
  def witaj(self):
    super().witaj()
    print("Witaj B")

class C(A):
  def witaj(self):
    super().witaj()
    print("Witaj C")

class D(B, C):
  def witaj(self):
    super().witaj()
    print("Witaj D")
```

```python
d = D()
d.witaj()
```
