```python
#Stwórz kalkulator który pobiera dwie wartości od użytkownika
# następnie użytkownik podaje jaką operacje art. przeprowadzić
# na końcu wyświetla się wynik
```

```python
liczba_1 = int(input("Podaj wartosc"))

znak = input("Podaj jaką chcesz przeprowadzić operacje\nPodaj znak + - / * ")

liczba_2 = int(input("Podaj wartosc"))

if znak == "+":
  print("wynik:",liczba_1 + liczba_2)

elif znak == "-":
  print("wynik:",liczba_1 - liczba_2)

elif znak == "*":
  print("wynik:",liczba_1 * liczba_2)

elif znak == "/":

  if liczba_2 == 0:
    print("Nie mozna dzielic przez zero")

  else:
    print("wynik:",liczba_1 / liczba_2)

else:
  print("Podany znak operacji jest niepoprawny")
```
