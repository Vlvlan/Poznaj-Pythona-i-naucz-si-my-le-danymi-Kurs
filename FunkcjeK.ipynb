```python
def przywitaj(imie, wiek, kolor_oczu="brązowego"):
  return f"Masz na imię {imie}, masz {wiek} lat i twoje oczy są {kolor_oczu} koloru"

print(przywitaj("Marek", "20"))
```

```python
print(przywitaj(wiek=20,imie="Marek",kolor_oczu="Niebieskiego"))
```

```python
def money(miesiac, *args):
  print(type(args))
  return f"W miesiącu {miesiac} udało ci się uzbierac {sum(args)} złotych"

print(money("luty", 100,200,50,20))
```

```python
def silnia(x):

  if x == 1 or x == 0:
    return 1

  return x * silnia(x - 1)

silnia(10)
```
