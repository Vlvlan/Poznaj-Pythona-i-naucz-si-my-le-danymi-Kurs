```python
class Cake:

  cake_pieces = 10

  def __init__(self, waga, smak, swieczki):
    self.waga = waga
    self.smak = smak
    self.swieczki = swieczki


  def eat(self, how_many):

    if self.cake_pieces < how_many:

      return f"Nie zostało tyle kawałków ciasta, jest tylko {self.cake_pieces}kawałków"

    else:

      self.cake_pieces = self.cake_pieces - how_many
      return f"Zostało {self.cake_pieces} kawałków do zjedzenia"

  def show(self):
    ciasto_v = f"""
    Waga tortu: {self.waga}kg
    Smak tortu: {self.smak}
    Ilość swieczek na torcie: {self.swieczki}"""

    return ciasto_v
```

```python
Cake.cake_pieces = 20
```

```python
ciasto_1 = Cake(1, "truskawka", "20")
```

```python
print(ciasto_1.show())
```

```python
class Order(Cake):

  def buy(self):

    money = 0

    if int(self.waga) > 1 and str(self.smak).lower() == "truskawka":
      money = 200

    elif int(self.waga) > 1 and str(self.smak).lower() == "czekolada":
      money = 180

    elif int(self.waga) > 1 and str(self.smak).lower() == "smietanka":
      money = 150
    else:
      money = 100

    return f"Koszt tortu to {money} złotych, twój tort to:\n{self.show()}"
```

```python
kup = Order(1, "truskawka", 20)
```

```python
print(kup.buy())
```

```python
kup_2  = Order(2, "czekolada", 20)
```

```python
print(kup_2.buy())
```
