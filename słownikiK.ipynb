```python
slownik = {"Chleb" : 5, "Burak" : 1, "Pieczony kurczak" : 6 }
print(slownik)
```

```python
print(slownik["Burak"])
```

```python
for i in slownik.keys():
  print("Produkty jadalne:", i)
```

```python
for i in slownik.values():
  print("Punkty głodu", i)
```

```python
for produkt, punkt in slownik.items():
  print(f"Produkt {produkt} daje {punkt} punktów głodu")
```

```python
slownik.get("Burak")
```

```python
print(slownik.get("arbuz", "Brak podanego klucza"))
```

```python
slownik.setdefault("Burak")
```

```python
slownik.setdefault("arbuz", 1)
print(slownik)
```

```python
slownik.pop("Burak")
```

```python
slownik.popitem()
```

```python
slownik_2  = {"Oko pająka" : 2, "ryba" : 2}
slownik.update(slownik_2)
print(slownik)
```

```python
polska_gura = {"Obrona" : {"Jan bednarek" : 20, "Jakub kiwior" : 15}, "Bramkarz" : {"Marcin Bułka": 10, "Kamil Grabara" : 60} }
print(polska_gura)
```

```python
print(polska_gura["Obrona"])
```

```python
print(polska_gura["Obrona"]["Jan bednarek"])
```

```python
for kategoria, wartosc in polska_gura.items():

  print("Kategoria", kategoria,"\n")

  for imie,numer in wartosc.items():
    print(f"Zawodnik {imie}, z numerek: {numer}")
```
