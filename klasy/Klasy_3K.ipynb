```python
class Pizza:

  def __init__(self, rozmiar, numer):
    self.rozmiar = rozmiar
    self.numer = numer

  def inna(self):
    return self.rozmiar

  @staticmethod
  def numer_pizzy():
    opcje = """
    1.Margherita - ser, mozzarella, oregano
    2.Capriciosa - ser, szynka, pieczarki
    3.Hawajska - ser, szynka, ananas
    """
    return opcje
```

```python
pizza = Pizza("Duza", 2)
```

```python
print(pizza.numer)
```

```python
print(Pizza.numer_pizzy())
```

```python
class Temperatura:

  def __init__(self, stopnie, pogoda):

    self.stopnie = stopnie
    self.pogoda = pogoda

  def __str__(self):
    return f"Dzisiaj temperatura wynosi {self.stopnie}, i jest {self.pogoda} pogoda."

  @staticmethod
  def zamiana(x):
    return x * 1.8 + 32
```

```python
day_1 = Temperatura(30, "słonecznie")
print(day_1)
```

```python
print(Temperatura.zamiana(30))
```

```python
print(day_1.zamiana(30))
```

```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age


  def show(self):
   print(f"Imie {self.name}, wiek {self.age}")


  @classmethod
  def metoda_klasy(cls, data):
    name, age = data.split(",")
    return cls(name, age)
```

```python
h1 = Human("Marek", 20)
```

```python
h2 = Human.metoda_klasy("Paweł,30")
```

```python
h1.show()
```

```python
h2.show()
```
