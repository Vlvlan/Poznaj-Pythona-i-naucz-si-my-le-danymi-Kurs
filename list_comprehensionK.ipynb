```python
l = []

for i in range(0,11):
  l.append(i)
print(l)
```

```python
lista = [x for x in range(0,11)]
print(lista)
```

```python
lista_2 = [x**2 for x in range(0,11)]
print(lista_2)
```

```python
lista_3 = [x for x in range(0,11) if x % 2 == 1]
print(lista_3)
```

```python
lista_4 = ["parzysta" if x % 2 == 0 else "nieparzysta" for x in range(0,11)]
print(lista_4)
```
