```python
nazwiska = ["Kowalski", "Wozniak", "Młot"]
egz = [50, 41, 69]
for n,e in zip(nazwiska,egz):
# f.zip() służy do łączenia elementów z dwóch lub więcej obiektów iterowalnych
# przy łączeniu dwóch obiektów zwraca wynik jako krotke czyli obiekt tuple
  print(f"Uczen {n} zdobył {e} punktów")
```
